```python
# Subplots of Distribution of Gender, Customer & Types of Productline, Payment Mode 
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,4,1)
ax=sns.countplot(data=df, x="Gender",hue="Gender",palette = 'pastel',legend=False)
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")

plt.subplot(1,4,2)
ax=sns.countplot(data=df, x='Customer type',hue="Customer type",palette = 'deep',legend=False)
ax.bar_label(ax.containers[0])
plt.title("Customer Distribution")
plt.show()

plt.figure(figsize=(18,6))
plt.subplot(1,4,3)
ax=sns.countplot(data=df,x='Product line',hue="Product line",palette = 'muted',legend=False)
ax.bar_label(ax.containers[0])
plt.title("Types of Product")
plt.xticks(rotation=45)

plt.subplot(1,4,4) 
sns.countplot(data=df,x='Payment',hue="Payment",palette = 'colorblind',legend=False)
plt.title("Payment Mode")


```

 # Product Line Analysis

```python

sns.countplot(data=df, x='Product line',hue='Gender')
plt.xticks(rotation=45)
plt.title('Product Line vs Gender')
plt.legend(title='Gender', loc='upper left', bbox_to_anchor=(-0.02, 1)) 
```

```python
sns.barplot(data=df,x="Product line",y="Rating", estimator='sum',palette = 'muted',hue="Product line" )
plt.xticks(rotation=45)
plt.title('Total Rating on Product line')
```

```python
df2=df.groupby("Product line")["Sales"].sum().reset_index()
df2
```

```python
sns.barplot(data=df2, x='Product line', y='Sales',palette = 'colorblind')
plt.title('Total Sales of Product line')
plt.xticks(rotation=45)
```

```python
sns.scatterplot(data=df,x="Unit price", y="cogs", alpha=0.5)
plt.title("Relationship Between Unit Price and COGS")
```

# City By Sales

```python
df1=df.groupby(["City"])["Sales"].sum().reset_index()
df1
```

```python
sns.barplot(data=df1,x="City",y="Sales",palette='Set3',hue="City",legend=False)
plt.title("Total Sales by City")
```

```python
sns.barplot(x='Payment', y='Sales', data=df,estimator=sum,palette = 'deep',hue="Payment",legend=False)
plt.title("Payment by total Sales")
```

```python
sns.barplot(data=df, x='City', y='Sales',hue='Branch',  estimator='sum')
plt.title("Branch-wise Sales Across Cities")
plt.show()

```

```python
sns.barplot(data=df, x='Customer type', y='Sales', hue='Gender', estimator='sum')
plt.title("Sales by Customer Type and Gender")
plt.show()

```
